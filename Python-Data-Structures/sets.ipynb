```python
set_a = {1,2,3,4}
set_b = {3,4,5,6}

print(set_a.union(set_b))
# print(set_a | set_b)

print(set_a.intersection(set_b))
# print(set_a & set_b)

print(set_a.difference(set_b))
# print(set_a - set_b)

print(set_a.symmetric_difference(set_b))
# print(set_a ^ set_b)
```
