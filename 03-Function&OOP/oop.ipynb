```python
class System:
    def __init__(self,name,price,quantity):
        self.name=name
        self.price=price
        self.quantity=quantity
    def restock(self, quantity):
        self.quantity += quantity
    def sell(self, quantity):
        if quantity> self.quantity:
            return "Flase"
        else:
            self.quantity -=quantity
    def display(self):
        print(f"Name: {self.name} , price: {self.price}, quantity: {self.quantity}")



s = System("Lenovo",45,90)

s.display()





```

```python
class Msys:
    def __init__(self, name , ID ):
        self.name=name
        self.ID=ID
        self.list=[]
    def add(self, list):
        self.list.append(list)
    def display(self):
        print(f"Name: {self.name}, ID: {self.ID}, list of courses: {self.list}")

s = Msys("abdo", 210421)

s.add("AI")
s.add("IOT")

s.display()
```

```python
class banksys:
    def __init__(self, Name , accountNum , address , balance):
        self.Name=Name
        self.accountNum=accountNum
        self.address=address
        self.balance=balance
    def deposite(self, balance):
        self.balance += balance
    def withdraw(self, balance):
        self.balance -= balance
    def display(self):
        print(f"Name: {self.Name}\nAccountNumber: {self.accountNum},\nAddress:{self.address}\nBalance: {self.balance}")    
    
bank = banksys("ahmed",48645,"Helwan",1200)

bank.display()
bank.withdraw(500)
bank.display()


```

```python
class Lib:
    def __init__(self, title="" , author=""):
        self.title=title
        self.author=author
        self.dict ={}
    def add( self , title , author):
        self.dict[author] ={title}
    def display(self):
        print(self.dict)
    
lib = Lib()
lib.add("book1","abdo")
lib.add("book2","Ahmed")
lib.display()
```

```python
class votingSys:
    def __init__(self,cnt=0 ,n = "", flag=""):
        self.cnt = cnt
        self.dict = {}
    def add(self, n , flag):
        self.dict[n]={flag}
        self.cnt +=1
    def displayNUM(self):
        print(self.cnt)
    def display(self):
        print(self.dict)

candidate = votingSys()

candidate.add("ahmed","Yes")
candidate.add("Mo","NO")

candidate.displayNUM()
candidate.display()
```
